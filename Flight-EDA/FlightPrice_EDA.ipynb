## **Performing Feature Engineering**

### **Handeling 'Date_of_Journey'**

```python
final_df['Date'] = final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month'] = final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year'] = final_df['Date_of_Journey'].str.split('/').str[2]
final_df.head(2)
```

```python
final_df = final_df.drop('Date_of_Journey', axis=1)
final_df['Date'] = final_df['Date'].astype(int)
final_df['Month'] = final_df['Month'].astype(int)
final_df['Year'] = final_df['Year'].astype(int)
final_df.info()
```

### **handeling of 'Arrival_Time'**

```python
final_df['Arrival_Time'] = final_df['Arrival_Time'].str.split(' ').str[0]
final_df.head(2)
```

```python
final_df['Arrival_hour'] = final_df['Arrival_Time'].str.split(':').str[0]
final_df['Arrival_min'] = final_df['Arrival_Time'].str.split(':').str[1]
final_df['Arrival_hour'] = final_df['Arrival_hour'].astype(int)
final_df['Arrival_min'] = final_df['Arrival_min'].astype(int)
final_df = final_df.drop('Arrival_Time', axis=1)
final_df.info()
```

### **Handeling of 'Dep_Time'**

```python
final_df['Dep_Time'] = final_df['Dep_Time'].str.split(' ').str[0]
final_df.head(2)
```

```python
final_df['Dep_hour'] = final_df['Dep_Time'].str.split(':').str[0]
final_df['Dep_min'] = final_df['Dep_Time'].str.split(':').str[1]
final_df['Dep_hour'] = final_df['Dep_hour'].astype(int)
final_df['Dep_min'] = final_df['Dep_min'].astype(int)
final_df = final_df.drop('Dep_Time', axis=1)
final_df.info()
```

### **Handeling of 'Total_Stops'**

```python
final_df['Total_Stops'].unique()
```

```python
final_df[final_df['Total_Stops'].isnull()]
```

```python
final_df['Total_Stops'] = final_df['Total_Stops'].map({'non-stop': 0, '1 stop': 1, 'Nan' : 1, '2 stops': 2, '3 stops':3, '4 stops':4})
final_df.drop('Route', axis=1, inplace=True)
final_df.head(2)
```

### **Handeling of 'Duration'**

```python
final_df['Duration_hour'] = final_df['Duration'].str.split(' ').str[0].str.split('h').str[0]
final_df['Duration_min'] = final_df['Duration'].str.split(' ').str[1].str.split('m').str[0]
final_df.head(2)
```

```python
final_df[final_df['Duration_hour'] == '5m']
```

```python
final_df.drop(6474, axis=0, inplace=True)
final_df.drop(2660, axis=0, inplace=True)
```

```python
final_df['Duration_hour'] = final_df['Duration_hour'].astype(int)
```

```python
final_df['Duration_min'] = final_df['Duration_min'].replace(np.nan, '0')
final_df['Duration_min'] = final_df['Duration_min'].astype(int)
final_df.drop('Duration', axis=1, inplace=True)
final_df.head(2)
```

### **Handeling of 'Airlines'**

```python
final_df['Airline'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
```

```python
final_df['Airline'] = label.fit_transform(final_df['Airline'])
final_df['Source'] = label.fit_transform(final_df['Source'])
final_df['Destination'] = label.fit_transform(final_df['Destination'])
final_df['Additional_Info'] = label.fit_transform(final_df['Additional_Info'])
final_df.head()
```
